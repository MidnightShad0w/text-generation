```python
from yandex_cloud_ml_sdk._auth import APIKeyAuth
from yandex_cloud_ml_sdk import YCloudML

folder_id = '******'
auth = '***********'

sdk = YCloudML(
    folder_id=folder_id,
    auth=APIKeyAuth(auth),
)

model = sdk.models.completions("yandexgpt")
```

Создание диалога

```python
messages = [
    {"role": "system", "text": "Ты — ассистент, который создает диалоги между персонажами видеоигр. \nВ ответе используй неожиданный сюжетный поворот в контексте заданного запроса."},
    {"role": "user", "text": "Создай интересный диалог между персонажами. Два героя обсуждают планы по спасению королевства."},
]

result = (
    model.configure(temperature=0.5).run(messages)
)

for alternative in result:
    print(alternative.text)
```

Использование fine-tuning для корректировки ответов модели.

В обучающем датасете всегда упоминается место действия - Врата Балдура

```python
import pathlib

def local_path(path: str) -> pathlib.Path:
    return pathlib.Path(path)
```

```python
dataset_path = local_path("tuning.jsonlines")
print(dataset_path)

print("Dataset:")
print(dataset_path.read_text(encoding='utf-8'))

dataset_draft = sdk.datasets.draft_from_path(
    task_type='TextToTextGeneration',
    path=dataset_path,
    upload_format='jsonlines',
    name='test-tuning'
)

dataset = dataset_draft.upload()
print(f'New dataset: {dataset=} \n')
```

```python
for dataset in sdk.datasets.list():
    print(f"List of existing datasets {dataset=}")
```

```python
# for dataset in sdk.datasets.list():
#     dataset.delete()
```

```python
import uuid

base_model = sdk.models.completions("yandexgpt-lite")
```

```python
tuning_task = base_model.tune_deferred(
    dataset,
    name=str(uuid.uuid4()),
    n_samples=6000
)

print(f'Start learning {tuning_task} \n')

new_model = tuning_task.wait()
print(f'Learning completed, new model = {new_model} \n')
```

Также пробовал n_samples=3000, но модель очень криво отвечает и чаще всего ссылается на поисковик яндекса. Точную причину назвать не могу, но предполагаю, что модель мало обучалась и качество датасета так себе.
